### Cleaning US_Per_Capita spending

```python
path_US=Path("../SpendingData/US_PER_CAPITA20.CSV")
US_df=pd.read_csv(path_US, encoding="UTF-8")
US_df.head()
```

```python
US_df.shape
```

```python
US_df['Item'].unique()
```

```python
US_Per_df=US_df.dropna(how="any")
US_Per_df.count()
```

```python
US_Per_df.shape
```

```python
US_Per_df.columns
```

```python
US_Per_df=US_Per_df[['Item', 'State_Name', 'Y2010', 'Y2011', 'Y2012', 'Y2013', 'Y2014',
       'Y2015', 'Y2016', 'Y2017']]

US_Per_df.head()
```

```python
US_Per_rename_df=US_Per_df.rename(columns={'State_Name':'State'})
US_Per_rename_df.head()                    
```

```python
years_list=['Y2010', 'Y2011', 'Y2012', 'Y2013', 'Y2014',
       'Y2015', 'Y2016', 'Y2017']
dfs=[]
for year in years_list:
    df=US_Per_rename_df[["Item", "State", year]]
    df=df.rename(columns={year:"Spending"})
    df["Year"] = int(year.replace("Y",""))
    dfs.append(df)
```

```python
US_Per_fixed_df=pd.concat(dfs)
```

```python
US_Per_fixed_df.head()
```

```python
shifting_pivot = pd.pivot_table(US_Per_fixed_df, values='Spending', index=['Year', 'State'], columns='Item')
shifting_pivot=shifting_pivot.reset_index()
shifting_pivot.head()
```

```python
shifting_pivot.columns
```

### Suming all data (.sum did not work)
