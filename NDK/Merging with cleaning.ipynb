# Opening other csv file_PerCap.csv (Clean_spending per person)

```python
csv_path = Path("./Clean_Per_Cap.csv")
PerCap_df = pd.read_csv(csv_path, encoding="UTF-8")
PerCap_df.head()
```

```python
PerCap_df.columns
```

# Reducing columns

```python
reduced_PerCap_df = PerCap_df[['Item','State_Name', 'Y2010', 'Y2011', 'Y2012', 'Y2013', 'Y2014',
       'Y2015', 'Y2016', 'Y2017', 'Y2018', 'Y2019', 'Y2020','Average_Annual_Percent_Growth']]
reduced_PerCap_df.columns
```

```python
reduced_PerCap_df.head()
```

# Looping through years and merging using pd.concat(). 
# Replacing name Y to nothing

```python
years_list=['Y2010', 'Y2011', 'Y2012', 'Y2013', 'Y2014',
       'Y2015', 'Y2016', 'Y2017', 'Y2018', 'Y2019', 'Y2020']
dfs=[]
for year in years_list:
    df=reduced_PerCap_df[["Item", "State_Name", year]]
    df=df.rename(columns={year:"Spending"})
    df["Year"] = int(year.replace("Y",""))
    dfs.append(df)
```

```python
result=pd.concat(dfs)
```

```python
result.to_csv("data.csv", index=False, header=True, encoding="UTF-8")
```

# Moving Year in the rows to the columns using .pivot_table()
# Column State_Name to State

```python
cols=['Year','State_Name', 'Dental Services ($)',
       'Hospital Care ($)', 'Personal Health Care ($)',
       'Physician & Clinical Services ($)', 'Home Health Care ($)']
```

```python
pivot_table = pd.pivot_table(result, values='Spending', index=['State_Name', 'Year'], columns='Item')
pivot_table=pivot_table.reset_index()
pivot_table=pivot_table[cols].rename(columns={"State_Name":"State"})
pivot_table
```

# Merge tables: Spending and Income by State and Year.

```python
final_result_df=pd.merge(filtered_Income_df, pivot_table, on=["Year", "State"], how="left")
final_result_df
```

# Save final_result_df (Final-merged table) as Final_table.csv

```python
final_result_df.to_csv("Final_table.csv", index=False, header=True, encoding="UTF-8")
```

# Reducing # of columns

```python
reduced_PHI_Per_df=PHI_Per_df[['Item', 'State_Name',
       'Y2001', 'Y2002', 'Y2003', 'Y2004', 'Y2005', 'Y2006', 'Y2007', 'Y2008',
       'Y2009', 'Y2010', 'Y2011', 'Y2012', 'Y2013', 'Y2014', 'Y2015', 'Y2016',
       'Y2017', 'Y2018', 'Y2019', 'Y2020', 'Average_Annual_Percent_Growth']]
reduced_PHI_Per_df.columns
```

# Looping though years and merging years on top of each other using .concat()

```python
years_list_2=['Y2010', 'Y2011', 'Y2012', 'Y2013', 'Y2014',
       'Y2015', 'Y2016', 'Y2017', 'Y2018', 'Y2019', 'Y2020']
dfs_2=[]
for year in years_list_2:
    df=reduced_PHI_Per_df[["Item", "State_Name", year]]
    df=df.rename(columns={year:"Spending"})
    df["Year"] = int(year.replace("Y",""))
    dfs_2.append(df)
```
